# Extract CellComplex from the imported cluster

```python
cells = Topology.Cells(objects[0])
print("Number of cells:", len(cells))

cc = CellComplex.ByCells(cells)
cc = Topology.RemoveCoplanarFaces(cc)
cc = Topology.RemoveCollinearEdges(cc)
print("CellComplex:", cc)
```

# Pass cc (CellComplex), NOT objects (list) to derive dual Graph



```python

g2 = Graph.ByTopology(cc, direct=True)
```

# Assign Visual attributes to edges and vertices

```python
vertices = Graph.Vertices(g2)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(g2)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

# Show Graph

```python
Topology.Show(g2,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```
