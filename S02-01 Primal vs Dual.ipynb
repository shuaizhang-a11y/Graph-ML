## 1. Import the needed classes

```python
from topologicpy.Vertex import Vertex
from topologicpy.Face import Face
from topologicpy.Shell import Shell
from topologicpy.Cell import Cell
from topologicpy.CellComplex import CellComplex
from topologicpy.Cluster import Cluster
from topologicpy.Topology import Topology
from topologicpy.Dictionary import Dictionary
from topologicpy.Graph import Graph
from topologicpy.Helper import Helper
```

## 2. Check the TopologicPy version

```python
print("This tutorial requires topologicpy version 0.9.18 or newer.")
print(Helper.Version())
```

## 3. Set your renderer:
* Visual studio code: "vscode"
* Google Colab: "colab"
* Browser: "browser"

```python
renderer = "vscode"
```

## 4. Create the geometry (Three rooms)

```python
r1 = Cell.Prism(origin=Vertex.ByCoordinates(0,0,0), width=5, length=7, height=3, placement="lowerleft")
r2 = Cell.Prism(origin=Vertex.ByCoordinates(5,0,0), width=3.5, length=3.5, height=3, placement="lowerleft")
r3 = Cell.Prism(origin=Vertex.ByCoordinates(5,3.5,0), width=3.5, length=3.5, height=3, placement="lowerleft")
house = CellComplex.ByCells([r1, r2, r3])
d1 = Face.Rectangle(origin=Vertex.ByCoordinates(7.5,3.5,0.05), direction=[0,-1,0], width=0.8, length=2.1, placement="lowerleft")
d2 = Face.Rectangle(origin=Vertex.ByCoordinates(5,0.2,0.05), direction=[-1,0,0], width=2.1, length=0.8, placement="lowerleft")
house = Topology.AddApertures(house, [d1, d2], subTopologyType="face")
```

## 5. Show the geometry

```python
Topology.Show(house, d1, d2,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 6. Derive primal graph

```python
cluster = Cluster.ByTopologies([house, d1, d2])
vertices = Topology.Vertices(cluster)
vertices = [Topology.Copy(v) for v in vertices]
edges = Topology.Edges(cluster)
edges = [Topology.Copy(e) for e in edges]
g1 = Graph.ByVerticesEdges(vertices, edges)
```

## 7. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g1)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(g1)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

## 8. Show the geometry and the graph

```python
Topology.Show(house, d1, d2, g1,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 9. Derive dual graph (based on cells in the cellcomplex)

```python
g2 = Graph.ByTopology(house, direct=True)
```

## 10. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g2)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(g2)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

## 11. Show the geometry and graph

```python
Topology.Show(house, d1, d2, g2,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 12. Derive dual graph based on the faces of a shell

```python
faces = Topology.Faces(house)
shell = Shell.ByFaces(faces)
g3 = Graph.ByTopology(shell, direct=True)
```

## 13. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g3)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(g3)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

## 14. Show the geometry and graph

```python
Topology.Show(house, d1, d2, g3,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
               width=500,
              height=500,
              backgroundColor="white",
              renderer=renderer)
```
