## Connect Internal Nodes and External Graph Nodes

This section demonstrates how to connect the manually created internal nodes graph with the automatically generated external graph from the building model. This enables a complete mapping from the building's interior space to the external topological structure.

```python
from topologicpy.Vertex import Vertex
from topologicpy.Edge import Edge
from topologicpy.Cluster import Cluster
from topologicpy.Topology import Topology

# Step 1: Extract all vertices and edges from the external building model graph
external_vertices = Graph.Vertices(g)
external_edges = Graph.Edges(g)

# Use all external vertices (or minimal sampling) to fill interior space with dense nodes
# Sample every 2nd vertex to balance density and performance
interior_vertices = external_vertices[::2] if external_vertices else []

# Step 2: Get all edges that connect sampled interior vertices
interior_edges = []
for edge in external_edges:
    v_start = Edge.StartVertex(edge)
    v_end = Edge.EndVertex(edge)
    if v_start in interior_vertices and v_end in interior_vertices:
        interior_edges.append(edge)

# Step 3: Create additional connecting edges between interior vertices to increase density
# Connect nearby vertices within the interior network
additional_interior_edges = []
for i in range(0, len(interior_vertices) - 1, 3):
    for j in range(i + 1, min(i + 6, len(interior_vertices))):
        additional_interior_edges.append(Edge.ByVertices([interior_vertices[i], interior_vertices[j]]))

# Step 4: Create edges connecting internal manual nodes and interior vertices
# This bridges the manually placed internal nodes with the automatically generated network
bridge_edges = []

for internal_node in internal_nodes:
    # Connect each internal node to the 5 nearest interior vertices
    for int_v in interior_vertices[::max(1, len(interior_vertices)//5)]:
        bridge_edges.append(Edge.ByVertices([internal_node, int_v]))

# Step 5: Create complete interior space graph with dense node and edge network
# Contains: model + interior nodes/edges + bridge edges connecting manual nodes
interior_graph_display = Cluster.ByTopologies([model_topology] + 
                                              interior_edges + 
                                              interior_vertices + 
                                              additional_interior_edges +
                                              internal_nodes +
                                              bridge_edges)

# Step 6: Display the dense interior network
# Red vertices and dark gray/black edges similar to the reference image
Topology.Show(
    interior_graph_display,
    faceColor=[200,200,200],       # Model face color (light gray)
    faceOpacity=0.15,              # Model opacity (very transparent)
    edgeColor="black",             # Edges are black/dark
    edgeWidth=1,                   # Thin edges for dense visibility
    vertexColor="red",             # Red vertices
    vertexSize=14,                  # Small vertex size for dense appearance
    showVertices=True,             # Show all vertices
    backgroundColor="white",       # White background
    width=1000,                    # Canvas width
    height=750,                    # Canvas height
    renderer=renderer
)

print(f"Interior vertices: {len(interior_vertices)}")
print(f"Interior edges: {len(interior_edges)}")
print(f"Additional interior edges: {len(additional_interior_edges)}")
print(f"Bridge edges: {len(bridge_edges)}")
print(f"Total edges in display: {len(interior_edges) + len(additional_interior_edges) + len(bridge_edges)}")
print(f"Total vertices in display: {len(interior_vertices) + len(internal_nodes)}")
```

## Connect Opposite Face Centers

This section creates nodes at the center of each face and connects opposite faces with edges, creating a spatial network that captures the relationship between opposing surfaces.

```python
from topologicpy.Vertex import Vertex
from topologicpy.Edge import Edge
from topologicpy.Face import Face
from topologicpy.Topology import Topology
from topologicpy.Cluster import Cluster
from topologicpy.Color import Color
import math

# Step 1: Extract all faces and calculate their center points
all_faces = Topology.Faces(cluster)
print(f"Total faces: {len(all_faces)}")

face_centers = []
for face in all_faces:
    center = Topology.CenterOfMass(face)  # Use Topology method instead of Face method
    face_centers.append((face, center))

# Step 2: Find opposite faces by maximum distance and create connecting edges
opposite_edges = []
connected_pairs = set()

for i, (face_i, center_i) in enumerate(face_centers):
    if i in [p[0] for p in connected_pairs] or i in [p[1] for p in connected_pairs]:
        continue
    
    max_distance = 0
    opposite_face_idx = -1
    
    for j, (face_j, center_j) in enumerate(face_centers):
        if i == j or j in [p[0] for p in connected_pairs] or j in [p[1] for p in connected_pairs]:
            continue
        
        dist_x = center_i.X() - center_j.X()
        dist_y = center_i.Y() - center_j.Y()
        dist_z = center_i.Z() - center_j.Z()
        distance = math.sqrt(dist_x**2 + dist_y**2 + dist_z**2)
        
        if distance > max_distance:
            max_distance = distance
            opposite_face_idx = j
    
    if opposite_face_idx >= 0:
        opposite_edges.append(Edge.ByVertices([center_i, face_centers[opposite_face_idx][1]]))
        connected_pairs.add((i, opposite_face_idx))

face_center_vertices = [center for _, center in face_centers]

print(f"Face center vertices created: {len(face_center_vertices)}")
print(f"Opposite connections created: {len(opposite_edges)}")

# Step 3: Create separate clusters for the two relationship types
interior_network = Cluster.ByTopologies(
    [model_topology] + 
    interior_edges + 
    interior_vertices
)

opposite_relationships = Cluster.ByTopologies(
    face_center_vertices + 
    opposite_edges
)

# Step 4: Display interior space network (red nodes, gray lines)
print("\n--- Interior Space Network ---")
Topology.Show(
    interior_network,
    faceColor=[200,200,200],
    faceOpacity=0.12,
    edgeColor="gray",
    edgeWidth=1,
    vertexColor="red",
    vertexSize=13,
    showVertices=True,
    backgroundColor="white",
    width=1000,
    height=750,
    renderer=renderer
)

# Step 5: Create combined view with face centers overlaid
# Layer the opposite relationships on top of interior network
combined_display = Cluster.ByTopologies(
    [model_topology] +           # Model
    interior_edges +             # Interior network edges (will show as gray)
    interior_vertices +          # Interior network vertices (will show as red)
    face_center_vertices +       # Face center vertices (will show with different symbol)
    opposite_edges               # Opposite connections (will show as thicker/different)
)

print("\n--- Combined View: Interior Network + Opposite Face Relationships ---")
Topology.Show(
    combined_display,
    faceColor=[210,210,210],
    faceOpacity=0.10,
    edgeColor=[80,80,80],        # Dark gray for interior edges
    edgeWidth=1,
    vertexColor="red",           # All vertices appear red in simple Show
    vertexSize=14,
    showVertices=True,
    backgroundColor="white",
    width=1000,
    height=750,
    renderer=renderer
)

print(f"\nTotal interior vertices: {len(interior_vertices)}")
print(f"Total face centers: {len(face_center_vertices)}")
print(f"Total interior edges: {len(interior_edges)}")
print(f"Total opposite edges: {len(opposite_edges)}")
```
