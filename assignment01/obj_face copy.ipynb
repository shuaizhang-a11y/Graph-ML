```python
cells = Topology.Cells(objects)
print("Number of cells:", len(cells) if cells else 0)

faces = Topology.Faces(objects)
print("Number of faces:", len(faces) if faces else 0)
```

```python
cc_building = None

if cells and len(cells) > 0:
    cc_building = CellComplex.ByCells(cells)
    print("CellComplex created from cells")
elif faces and len(faces) > 0:
    cc_building = CellComplex.ByFaces(faces)
    print("CellComplex created from faces")
else:
    print("No valid cells or faces found")
```

```python
if cc_building:
    Topology.Show(cc_building,
                  faceColor=[250,220,180],
                  faceOpacity=0.8,
                  edgeColor="white",
                  edgeWidth=2,
                  showVertices=False,
                  backgroundColor="black",
                  width=800,
                  height=600,
                  renderer=renderer)
```

```python
g = Graph.ByTopology(cc_building, direct=False)
print("Graph created")
print(g)
```

```python
g = Graph.ByTopology(cc_building, direct=False, viaSharedApertures=True)
```

```python
graph_vertices = Graph.Vertices(g)
graph_edges = Graph.Edges(g)

print("Number of graph vertices:", len(graph_vertices) if graph_vertices else 0)
print("Number of graph edges:", len(graph_edges) if graph_edges else 0)
```

```python
Topology.Show(g,
              vertexColor="red",
              vertexSize=6,
              edgeColor="lime",
              edgeWidth=2,
              backgroundColor="black",
              width=800,
              height=600,
              renderer=renderer)
```

```python
print(Topology.Analyze(objects))

cells = Topology.Cells(objects)
print("Number of cells:", len(cells) if cells else 0)

faces = Topology.Faces(objects)
print("Number of faces:", len(faces) if faces else 0)

cc_building = None

if cells and len(cells) > 0:
    cc_building = CellComplex.ByCells(cells)
    print("CellComplex created from cells")
elif faces and len(faces) > 0:
    cc_building = CellComplex.ByFaces(faces)
    print("CellComplex created from faces")
else:
    print("No valid cells or faces found")

Topology.Show(cc_building,
              faceColor=[250,220,180],
              faceOpacity=0.8,
              edgeColor="white",
              edgeWidth=2,
              showVertices=False,
              backgroundColor="black",
              width=800,
              height=600,
              renderer=renderer)

g = Graph.ByTopology(cc_building, direct=False)

print("Graph created")
print(g)

graph_vertices = Graph.Vertices(g)
graph_edges = Graph.Edges(g)

print("Number of graph vertices:", len(graph_vertices) if graph_vertices else 0)
print("Number of graph edges:", len(graph_edges) if graph_edges else 0)

Topology.Show(g,
              vertexColor="red",
              vertexSize=6,
              edgeColor="lime",
              edgeWidth=2,
              backgroundColor="black",
              width=800,
              height=600,
              renderer=renderer)
```
