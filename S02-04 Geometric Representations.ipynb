## 1. Import the needed classes

```python
from topologicpy.Cell import Cell
from topologicpy.Topology import Topology
from topologicpy.Dictionary import Dictionary
from topologicpy.Graph import Graph
from topologicpy.Helper import Helper
```

## 2. Check the TopologicPy version

```python
print("This tutorial requires topologicpy version 0.9.18 or newer.")
print(Helper.Version())
```

## 3. Set your renderer:
* Visual studio code: "vscode"
* Google Colab: "colab"
* Browser: "browser"

```python
renderer = "vscode"
```

## 4. Create the geometry (four walls and a floor)

```python
w1 = Cell.Prism(width=5, length=0.2, height=3)
w1 = Topology.Translate(w1, 0, 2.5, 0)
w2 = Topology.Translate(w1, 0, -5, 0)
w3 = Cell.Prism(width=0.2, length=5.2, height=3)
w3 = Topology.Translate(w3, 2.6, 0, 0)
w4 = Topology.Translate(w3, -5.2, 0 , 0)
f = Cell.Prism(width=5.4, length=5.2, height=0.2)
f = Topology.Translate(f, 0, 0, -1.6)
```

## 5. Show the geometry

```python
Topology.Show(w1, w2, w3, w4, f,
              width=500,
              height=500,
              backgroundColor="white",
              renderer=renderer
              )
```

## 6. Derive the graph from adjacency conditions

```python
g1 = Graph.BySpatialRelationships([w1, w2, w3, w4, f], include=["touches"])
```

## 7. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g1)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(g1)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

## 8. Show the geometry and the graph

```python
Topology.Show(w1, w2, w3, w4, f, g1,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              width=500,
              height=500,
              backgroundColor="white",
              renderer = renderer)
```

## 9. Illustrate intersection (clash detection)

```python
pipe = Cell.Cylinder(placement="bottom", direction=[1,0,0], radius=0.1, height=8)
pipe = Topology.Translate(pipe, -4, 1, 1.2)
```

## 10. Show the geometry

```python
Topology.Show(w1, w2, w3, w4, f, pipe,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              width=500,
              height=500,
              backgroundColor="white",
              renderer=renderer)
```

## 11. Derive intersection graph

```python
g2 = Graph.BySpatialRelationships([w1, w2, w3, w4, f, pipe], include=["overlaps"])
```

## 12. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g2)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(g2)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

## 13. Show the result

```python
Topology.Show(w1, w2, w3, w4, f, pipe, g2,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              width=500,
              height=500,
              backgroundColor="white",
              renderer=renderer)
```
